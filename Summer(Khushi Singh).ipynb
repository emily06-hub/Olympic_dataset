### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df
```

```python
sol = df.groupby(['Medal']).get_group('Gold')
```

```python
sol
```

```python
data = []

for Sport in sol['Sport'].unique():
  data.append([Sport , len(sol[sol['Sport'] == Sport])])

pd.DataFrame(data, columns = ['Sport','freq']).sort_values(by = 'freq', ascending = False).head()
```

```python
pd.DataFrame(data, columns = ['Sport','freq']).sort_values(by = 'freq', ascending = False).head().plot(x = 'Sport', y = 'freq', kind = 'bar')
```

### 6. In which year India won first Gold Medal in Summer Olympics?
